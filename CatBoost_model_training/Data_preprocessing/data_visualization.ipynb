```python
import pandas as pd
data=pd.read_csv("data.csv")
```

```python
data
```

```python
data["engine_size"]=data["engine_size"].astype(str)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
corr = data.select_dtypes(include="number").corr()
plt.figure(figsize=(10,8))
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(16,10))
sns.countplot(x="fuel",data=data,hue="fuel")
plt.title("Number of vehicules by fuel Type")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(16,10))
sns.countplot(x="transmission",data=data,hue="transmission")
plt.title("Number of vehicules by transmission")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(16,10))
sns.countplot(x="body_type",data=data,hue="body_type")
plt.title("Number of vehicules by body_type")
```

```python
n = 1000
counts = data["make"].value_counts()

models_rares = counts[counts >n].index

data = data[data["make"].isin(models_rares)]
```

```python
cars_by_Make=data["make"].value_counts().sort_values(ascending=False).reset_index(name="count")
plt.figure(figsize=(16,10))
plt.pie(x=cars_by_Make["count"],labels=cars_by_Make["make"],autopct="%1.2f%%")
plt.title("cars by make")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(16,10))
sns.countplot(x="drive_train",data=data,hue="drive_train")
plt.title("Number of vehicules by drive_train")
```

```python
data.to_csv("training_data.csv",index=False)
```
