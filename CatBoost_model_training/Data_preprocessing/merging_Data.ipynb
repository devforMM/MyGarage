```python
colonnes = [
    'price',
    'year',
    'mileage/mileage',
    'specification/make',
    'specification/model',
    'specification/trim',
    'specification/fuel',
    'specification/transmission',
    'specification/bodyType',
    'specification/doors',
    'specification/seats',
    'specification/engine/sizeLitres',
    'specification/engine/power/enginePower',
    'specification/driveTrain',
    'specification/co2Emissions/co2Emission',
    'specification/emissionClass',
    'colour',
    'specification/techData/fuelConsumptionCombined',
]
```

```python
import pandas as pd
import os
dfs=[]
for i,dataset in enumerate(os.listdir(".")) :
    try:
        data=pd.read_csv(dataset)
        dfs.append(
         data[colonnes]
        )
        
    except Exception as e:
        print(f"erreur : {e} dans le {dataset}")
```

```python
merged_data=pd.concat(dfs)
merged_data.columns=[
    "price",
    "year",
    "kilometres",
    "make",
    "model",
    "trim",
    "fuel",
    "transmission",
    "body_type",
    "doors",
    "seats",
    "engine_size",
    "engine_power",
    "drive_train",
    "co2_emission",
    "emission_class",
    "color",
    "fuel_consumption(L/100km)"
]
```

```python
merged_data.dropna(inplace=True)
```

```python
def convert(line):
    try:
     digit=line.split()[0]
     return 235.214/ float(digit)
    except Exception:
       return line
def convert_2(line):
    try:
     if line=="Golf Plus":
        return "Golf"
     else:
        return line
    except :
       pass

```

```python
merged_data["fuel_consumption(L/100km)"]=merged_data["fuel_consumption(L/100km)"].apply(convert)
merged_data["kilometres"]=merged_data["kilometres"]* 1.60934
merged_data["model"]=merged_data["model"].apply(convert_2)
```

```python
merged_data["fuel_consumption(L/100km)"]=merged_data["fuel_consumption(L/100km)"].round(0)
merged_data["kilometres"]=merged_data["kilometres"].round(0)
```

```python
merged_data["engine_size"]=merged_data["engine_size"].astype(str)
```

```python
def convert_hp(line):
    if type(line)==str:
     if line.isdigit():
        return float(line)
     else:
        return float(line[:2])
    elif type(line)==int:
       return float(line)
    elif type(line)==float:
       return float(line)
       
       
```

```python
merged_data["engine_power"]=merged_data["engine_power"].apply(convert_hp)
```
